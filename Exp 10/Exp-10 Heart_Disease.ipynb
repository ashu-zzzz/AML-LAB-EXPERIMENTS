```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Step 3: Load dataset
df = pd.read_csv("C:\\Users\\Shikhar Khanna\\Downloads\\AML lab\\heart.csv.xls")   # change name if needed

# Step 4: Split features & target
X = df.drop('target', axis=1)
y = df['target']

# Step 5: Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 6: Feature Extraction (PCA)
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X_scaled)

# Step 7: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=42
)

# Step 8: Models

# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)

# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

# SVM
svm = SVC()
svm.fit(X_train, y_train)
pred_svm = svm.predict(X_test)

# Step 9: Results
print("Logistic Regression:", accuracy_score(y_test, pred_lr))
print("Random Forest:", accuracy_score(y_test, pred_rf))
print("SVM:", accuracy_score(y_test, pred_svm))
```
